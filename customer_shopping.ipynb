# Análisis de datos

##ETL (Extracción, Transformación y Carga)

###Extracción

```python
from google.colab import drive
import pandas as pd
# Importación de una libreria que detecta el tipo de encoding
import chardet
# Montamos Google Drive para acceder al archivo
drive.mount('/content/drive', force_remount=True)
#
ruta_archivo = '/content/drive/MyDrive/Colab Notebooks/customer_shopping_data.csv'
with open(ruta_archivo, 'rb') as f:
    result = chardet.detect(f.read())

try:
  df = pd.read_csv(ruta_archivo, encoding=result['encoding'], sep=',', decimal=',')
  print('Extracción de datos exitosa')
except FileNotFoundError:
  print(f'El archivo no se encontró en la ruta: {ruta_archivo}')
```

###Transformación

```python
#Muestra las primeras 10 filas del Dataframe
df.head(10)
```

```python
#Resumen de la información del DataFrame
df.info()
```

```python
# Generar estadísticas descriptivas de las columnas numéricas
df.describe()
```

```python
# Renombrar columnas
df.rename(columns={
    'invoice_no' : 'factura_no',
    'customer_id' : 'cliente_id',
    'gender' : 'genero',
    'age' : 'edad',
    'category' : 'categoria',
    'price' : 'precio',
    'quantity' : 'cantidad',
    'payment_method': 'metodo_pago',
    'invoice_date': 'fecha_factura',
    'shopping_mall': 'centro_comercial'
}, inplace=True)

# Convertir la columna 'precio' a numérica
df['precio'] = df['precio'].astype(str)  # Primero lo convierto a string
df['precio'] = df['precio'].str.replace(',','.', regex=True).astype(float).round(2) # Reemplazo la coma por el punto

# Simplificar la columna 'genero'
df['genero']= df['genero'].str.replace('Male','M', regex=False).str.replace('Female','F', regex=False)

# Convertir la columna 'fecha_factura' al tipo datatime
df['fecha_factura'] = pd.to_datetime(df['fecha_factura'], format='%d/%m/%Y')

print ('Transformación de datos exitosa')
```

##Carga

```python
# Definir la ruta de salida para guardar el nuevo CSV
ruta_salida = '/content/drive/MyDrive/Colab Notebooks/customer_shopping_data_OK.csv'

# Guardar el DataFrame en un nuevo archivo CSV
df.to_csv(ruta_salida, index=False, sep=',', encoding = result['encoding'], decimal = '.')

print(f'DataFrame guardado en {ruta_salida}')
```
